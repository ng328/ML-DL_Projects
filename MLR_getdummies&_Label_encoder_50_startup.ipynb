```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("/content/sample_data/50_Startups (1).csv")
```

```python
dataset.shape
```

```python
dataset.dtypes
```

```python
dataset.isnull().any(axis=0)
```

```python
X=dataset.iloc[:,:-1]
```

```python
X
```

```python
Y=dataset.iloc[:,4]
```

```python
Y
```

```python
#convert the column into categorical  columns-->
states=pd.get_dummies(X['State'],drop_first=True)
```

```python
#https://www.statology.org/pandas-get-dummies/
```

```python
states
```

```python
#drop the state column
X=X.drop('State',axis=1)
```

```python
X
```

```python
#concat the dummy variables
X=pd.concat([X,states],axis=1)
```

```python
X
```

```python
#splitting the datasets into the training set and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
#fitting  MLR to the training set
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
#predicting the test set results
y_pred=reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
```

```python
score
```

```python
pd.DataFrame(zip(y_pred,y_test))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
from sklearn.linear_model import LinearRegression
df=pd.read_csv("/content/sample_data/50_Startups (1).csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab_enc=LabelEncoder()
```

```python
df.State=lab_enc.fit_transform(df.State)
```

```python
df.State
```

```python
df.State.unique()
```

```python
x=df.drop(['Profit'],axis=1)
```

```python
print(x.head())
```

```python
print(x.shape)
```

```python
y=df['Profit']
```

```python
print(y.head())
```

```python
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=45)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
```
